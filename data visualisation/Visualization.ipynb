# Types of data plotting

1.Numerical data plotting


```python
tips=sns.load_dataset('tips')
tips.tail()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)


#relplot-relational plot-find the relation between one data and the another
#here we are establishing the relationship with the bill and the tip
```

```python
#finding the number of smokers who did tip and did not tip
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='smoker',style='time')

#hue-gives different color for different values
#style-lunch and dinner is there in data and different styles given
```

```python
sns.relplot(x='total_bill',y='tip',  style='size',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, size='size')
```

2.Categorical dataplotting-two

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
```

# distPlot

```python
x=randn(1000) #x has random numbers
sns.distplot(x,kde=False,hist=False)
```

```python
#Shade inside the kde plot
sns.kdeplot(x,shade=True)  
```

# JOINTPLOT

```python
#joint will join two sets of data 
tips=sns.load_dataset('tips')
tips.tail()
x=tips['total_bill']
y=tips['tip']
sns.jointplot(x=x,y=y)
```

# REGRESSION PLOT

TYPES
1.regplot
2.lmplot


```python
#regplot

tips=sns.load_dataset('tips')
sns.regplot(x='total_bill',y='tip',data=tips)


#best fit line is the line that covers the maximum data
```

```python
#lmplot
sns.lmplot(x='total_bill',y='tip',data=tips)
```
