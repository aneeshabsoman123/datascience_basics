```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lm=LinearRegression()  #intialising
```

```python
df=pd.read_csv('Advertising.csv')

x=df[['TV','radio','newspaper']]  #feature
y=df.sales  #label

lm.fit(x,y)

print(lm.intercept_)
```

```python
print(lm.coef_[0])

#if one unit of TV advertisement is increased, then here 0.045 sales will get increased
```

```python
print(lm.coef_[1])

#if one unit of radio advertisement is increased, then here 0.045 sales will get increased

#if this becomes negative, that means the sales will go down
```

```python
print(lm.coef_[2])

#if one unit of newspaper advertisement is increased, then here 0.045 sales will get increased
```

```python
 import statsmodels.formula.api as smf

lm=smf.ols(formula='sales~TV +radio+newspaper',data=df).fit()
lm.conf_int()
lm.summary()
```
